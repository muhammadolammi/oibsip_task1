```python
import pandas as pd
```

```python
raw_data = pd.read_csv("iris.csv")
print(raw_data.shape)
raw_data.head()
```

```python
raw_data.describe()
```

```python
## We are dealing with classification so logit regression is a good solution/algorithm 
```

```python
df = raw_data.copy()
df.columns
```

```python
y = df["Species"]
features = [ 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
X = df[features]
```

```python
## we can map our y based on their species, we have 3 species so 0,1,2
print(y.unique())
y = y.map({"Iris-setosa":0, "Iris-versicolor":1, "Iris-virginica":2 })
```

```python
y.tail()
```

## Scaling Inputs

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
```

### split data into test and train

```python
from sklearn.model_selection import train_test_split
y_train, y_test, X_train, X_test = train_test_split(y, X_scaled, test_size=0.2, random_state=1)
```

### declare and fit/train the model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```


### get model parameters


```python
print("ACCURACY: ", model.score(X_train, y_train))
## The intercept will be 3 since we are classifying 3 classes, if it was 2 should be one
print("INTERCEPT: ", model.intercept_)
print("COEFFICIENTS: ", model.coef_)
```

## test  the model with the train data

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("ACCURACY: ", accuracy_score(y_test,y_pred ))
```
